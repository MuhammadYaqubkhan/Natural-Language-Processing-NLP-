## Sentence Tokenization

```python
import nltk
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(data)
sentences
```

```python
for sent in sentences:
    print(sent)
```

## Word Tokenization

```python
from nltk import word_tokenize
```

```python
listt = []
for sentence in sentences:
    words = word_tokenize(sentence)
    for w in words:
         listt.append(w)
```

```python
print(words)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
ver = []

lemmatizer = WordNetLemmatizer()

for word in listt:
    lemma = lemmatizer.lemmatize(word)
    ver.append(lemma)
    print(lemma)
```

## Stop words

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
sen = data

stop_words = set(stopwords.words("english"))    # We'll be setting English language

words = nltk.word_tokenize(sen)       # First splitting our sentence into words

without_stop_words = []              # Let's initialize an empty list where we'll be storing all the words except stop words
 
for word in words:                          # this will go through all the words one by one
    if word not in stop_words:              # If a word is not presesnt in the list of stopwords (predefined list of stopwords) 
        without_stop_words.append(word)     # append that word in our list of words without stopwords 
        
print(without_stop_words)                   # printing our list of words (without stop words)
print()
```

```python
senty = nltk.sent_tokenize(data)
print(senty)
print()
for word in senty:
    wordy = nltk.word_tokenize(word)
    print(wordy)
```

## Count Vectorizaton

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

bag_of_words = vectorizer.fit_transform(senty)

feature_names = vectorizer.get_feature_names()
feature_names
```

## Bag of words

```python
import pandas as pd
feature = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
feature
```
